```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf 
from tensorflow.keras.models import load_model
```

```python
model1=load_model('model_best.h5')
```

```python
class_labels= ['SUV', 'bus', 'family sedan', 'fire engine', 'heavy truck', 'jeep', 'minibus', 'racing car', 'taxi', 'truck']
```

```python
path = 'vehicle-classification/inference/0a5067c35b854ce7213e433d12ea500d.jpg'
```

```python

def prediction(file):
    img = tf.keras.utils.load_img(file, target_size=(150, 150))
    x = tf.keras.utils.img_to_array(img) / 255
    x = np.expand_dims(x, axis=0)

    images = np.vstack([x])
    classes = model1.predict(images, batch_size=10)
    idx = np.argmax(classes)

    plt.imshow(img)
    plt.show()  # Use plt.show() to display the image in a non-interactive environment

    predicted_class = class_labels[idx]
    predicted_proba = classes[0, idx]

    print('Prediction is a {} with probability: {:.2f}'.format(predicted_class, predicted_proba))

    # Print class names and their probabilities
    for class_label, probability in zip(class_labels, classes[0]):
        print('{}: {:.2f}'.format(class_label, probability))
```

```python
inf= prediction(path)
```
